```python
#Import nltk and 'punkt' to used the .tokenized() method
import nltk
nltk.download('punkt')
import pandas as pd
import numpy as np
```

```python
#Remove NAN
df = data[data['What_can_SAI_make_easier_for_you__c'].notna()]
df.head()
```

```python
#Remove NAN
df = df[df['Additional_Comments__c'].notna()]
df.head()
```

```python
from nltk.tokenize import word_tokenize
```

```python
#Use tokenizer to divide a text into a list of sentences using an unsupervised algorithm 
#TreebankWordTokenizer uses regular expressions to tokenize text and assumes the text has already been segmented into sentences
#RegexpTokenizer splits a string into substrings using a regular expression - used this method because it removes punctuations
from nltk.tokenize import RegexpTokenizer

tokens_list=[]

tokenizer = RegexpTokenizer(r'\w+')
for idx, row in df.iterrows(): 
#     if row['What_can_SAI_make_easier_for_you__c']: 
    tokens=tokenizer.tokenize(str(row['What_can_SAI_make_easier_for_you__c']))
    tokens_list.append(tokens)
```

```python
#Tokeniztion - A large quality of text divided into smaller parks called tokens.
#I.e. each row in index 10 was tokenized, the input (sentence?) was taken and then divided into each word
#Text has been converted to a list of lists
tokens_list
```

```python
#Remove stopwords from tokens_list
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

tokens1_stop = []
for i in range(2768):
    for word in tokens_list[i]:
        if word not in stop_words:
            tokens1_stop.append(word)
```

```python
#Impoert Lemmatizer
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
#Lemmatize tokens1_stop
lemmatized_output = [lemmatizer.lemmatize(w) for w in tokens1_stop]
print(lemmatized_output)
```

```python
#Lemmatize with POS Tag
from nltk import pos_tag
from nltk.corpus import wordnet
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
#POS Tag = Parts of Speech Tagging is a process to mark up the words in text format
tagged = pos_tag(tokens1_stop)
print(tagged)
```
