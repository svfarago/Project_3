```python
import pandas as pd

#Convert xlsx file to csv first, you may also have to resave a CSV file for the encoding (UTF-8)
df = pd.read_csv(r'RT_CLEAN_DATA.csv')
df.head()
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
#NLTK = Natural Language Tool Kit
#VADER = *not darth* Valence Aware Dictionary for Sentiment Reasoning
# -1 (most extreme negative) and +1 (most extreme positive)
```

```python
df.fillna("blank", inplace = True)
df.head(3)
```

```python
#----RESPONSE TEXT----
# Use iloc to start at row 0, looking at "Response Text Long"
# This is aggregate of all responses for neg, neu, pos, compound
sid.polarity_scores(df.iloc[0]['Response Text Long'])

```

```python
#----RESPONSE TEXT----
df['response_text_scores'] = df['Response Text Long'].apply(lambda Impediments:sid.polarity_scores(Impediments))

df['response_text_negative'] = df['response_text_scores'].apply(lambda a:a['neg'])
df['response_text_neutral'] = df['response_text_scores'].apply(lambda b:b['neu'])
df['response_text_positive'] = df['response_text_scores'].apply(lambda c:c['pos'])
df['response_text_compound'] = df['response_text_scores'].apply(lambda d:d['compound'])


df.head(3)
```

```python
df.to_csv(r'RT_VADER_CLEAN_DATA.csv')
```
