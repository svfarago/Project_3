```python
import pandas as pd

#Convert xlsx file to csv first, you may also have to resave a CSV file for the encoding (UTF-8)
df = pd.read_csv(r'2021 02 25 - Sentiment Analysis challenge NPS Feedback dataset.csv')
df.head()
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
#NLTK = Natural Language Tool Kit
#VADER = *not darth* Valence Aware Dictionary for Sentiment Reasoning
```

```python
df.fillna("blank", inplace = True)
df.head()
```

```python
sid.polarity_scores(df.iloc[0]['Additional_Comments__c'])
```

```python
sid.polarity_scores(df.iloc[0]['What_can_SAI_make_easier_for_you__c'])
```

```python
df['scores_additional'] = df['Additional_Comments__c'].apply(lambda Impediments:sid.polarity_scores(Impediments))
df['compound_additional'] = df['scores_additional'].apply(lambda d:d['compound'])
df['score_additional'] = df['compound_additional'].apply(lambda score: 'pos' if score >=0 else 'neg')
df.head()
```

```python
df['scores_for_you'] = df['What_can_SAI_make_easier_for_you__c'].apply(lambda Impediments:sid.polarity_scores(Impediments))
df['compound_for_you'] = df['scores_for_you'].apply(lambda d:d['compound'])
df['score_for_you'] = df['compound_additional'].apply(lambda score: 'pos' if score >=0 else 'neg')
df.head()
```

```python
df.to_csv(r'2021 03 26 - Sentiment Output.csv')
```
