```python
import pandas as pd

#Convert xlsx file to csv first, you may also have to resave a CSV file for the encoding (UTF-8)
df = pd.read_csv(r'RT_Responses_dataset_May2021.csv')
df.head()
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
#NLTK = Natural Language Tool Kit
#VADER = *not darth* Valence Aware Dictionary for Sentiment Reasoning
```

```python
df.fillna("blank", inplace = True)
df.head()
```

```python
# Just replaced original df with "Question Text"
df['question_text_scores'] = df['Question Text'].apply(lambda Impediments:sid.polarity_scores(Impediments))
df['question_text_compound'] = df['question_text_scores'].apply(lambda d:d['compound'])
df['question_text_scores'] = df['question_text_compound'].apply(lambda score: 'pos' if score >=0 else 'neg')
df.head()
```

```python
# Just replaced original df with "Response Text Long"
df['response_text_scores'] = df['Response Text Long'].apply(lambda Impediments:sid.polarity_scores(Impediments))
df['response_text_compound'] = df['response_text_scores'].apply(lambda d:d['compound'])
df['response_text_scores'] = df['response_text_compound'].apply(lambda score: 'pos' if score >=0 else 'neg')
df.head()
```

```python
df.to_csv(r'RT Sentiment Output_May2021.csv')
```
